Data voorbereiden

Importeer Pandas

```
import pandas as pd
```

Importeer bestand

```
df = pd.read_csv("./Sales_Data/Sales_July_2019.csv")

print(df.head())
```

Importeer operating system library

```
import os
```

Maak lijstje van bestanden in de map

```
path = "./Sales_Data/" 

files = [file for file in os.listdir(path)]

print(files)
```

Check!

```
for file in files:
    print(file)
```

Maak een dataframe met alle maanden erin

```
# maak lege dataframe aan om te vullen
all_months_data = pd.DataFrame()

# for loop om alle bestanden langs te gaan
for file in files: 
    current_data = pd.read_csv(path + "/" + file) # lees het bestand in het pad
    all_months_data = pd.concat([all_months_data, current_data])

print(all_months_data)
```

Sla dataframe op als nieuw bestand

```
all_months_data.to_csv("all_data.csv", index = False) # zonder de index van de losse bestanden
```

Open nieuwe bestand

```
all_data = pd.read_csv("all_data.csv") # lees bestand

print(all_data.head())
print(all_data.tail())
```

## 1 In welke maand werd het meest verkocht? Wat was de omzet in die maand?

Bekijk welke vorm en hoeveel dimensies je dataset heeft

```
print(all_data.shape)
print(all_data.ndim)
```

Nieuwe kolom: df["Ik ben een nieuwe kolom"]

```
# nieuwe kolom
                    # van de Order Date kolom
                                           # neem eerste twee tekens
all_data["Month"] = all_data["Order Date"].str[0:2]

print(all_data.head())
```

Bekijk en verwijder NaN

```
nope = all_data[all_data.isna().any(axis=1)]

print(nope.head(50))
```

```
all_data = all_data.dropna(how = "all") # dataset = dataset minus de NaN

print(all_data)
```

```
# verander data frame naar eentje zonder rijen met 'Or' mbv !=
all_data = all_data[all_data['Month'].str[0:2] != 'Or'] 

print(all_data)
```

Zet cijfers van maanden om in int

```
all_data["Month"] = all_data["Month"].astype("int32")
```

```
print(all_data.head(50))
```

Okay, nu zijn de waarden voor de 'maanden' as klaar om gebruikt te worden, op naar de omzet

Bereken bestelling prijs

Zet kolommen met cijfers om naar echte cijfers

```
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
```

```
all_data["Total"] = all_data["Quantity Ordered"] * all_data["Price Each"]

print(all_data.head())
```

Maak een variabele waar je per maand (groupby Month sum) bekijkt wat de omzet was (Total)

```
results_month = all_data.groupby("Month").sum()["Total"]

print(results_month)
```

Bekijk welke maand de meeste omzet had

```
print(results_month.sort_values(ascending=False))
```

Plot de info

```
import matplotlib.pyplot as plt
```

ðŸ”Ž matplotlib ticker millions

```
# dit van het internet gehaald
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

formatter = FuncFormatter(millions)

fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(formatter)

# maak x as eenheden
months = range(1, 13) # alle maanden, 1t/m12

# plot data
plt.bar(months, results_month) # waarden meegeven
plt.xticks(months) 
# plt.yticks()
plt.xlabel("Maanden")
plt.ylabel("Omzet in USD ($)")

plt.show()
```

## 2 Welke stad had meeste omzet?

```
print(all_data.head())
```

```
def get_city(address):
    return address.split(", ")[1]

def get_state(address):
    return address.split(", ")[2].split(" ")[0]
                                                                # vb New York     (NY)
all_data["City"] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

print(all_data.head())
```

Groepeer stad en geef omzet bij die stad

```
results_city = all_data.groupby("City", sort = False).sum()['Total']

print(results_city)
```

```
cities = all_data["City"].unique()

print(cities)
```

```
fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(formatter)

plt.bar(cities, results_city) # waarden meegeven
plt.xticks(cities, rotation = 90) 
# plt.yticks()
plt.xlabel("Steden")
plt.ylabel("Omzet in USD ($)")

plt.show()
```

## 3 Op welk tijdstip moeten we inzetten op adverteren?

```
# print(all_data["Order Date"])

all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])

all_data["Hour"] = all_data["Order Date"].dt.hour

hours = all_data["Hour"].sort_values().unique()

results_hours = all_data.groupby(["Hour"]).count()

plt.grid() # raster voor duidelijk

plt.xticks(hours)

plt.xlabel("Tijd")
plt.ylabel("Aantal bestellingen")

plt.plot(hours, results_hours)
```

## 4 Welk product werd het meest verkocht? En waarom?

Groeppeer obv product

```
results_product = all_data.groupby("Product").sum()["Quantity Ordered"]

print(results_product)
```

```
print(results_product.sort_values(ascending=False))
```

Plotten!

```
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = all_data["Product"].sort_values().unique()
print(keys)

plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

Plot mÃ©t prijs!

```
# vind dezelfde producten (groupby) en geef gemiddelde prijs van een product / groep (mean + Price Each)
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx() # deel samen een as
ax1.bar(keys, quantity_ordered, color='g') # de een is een bar chart,
ax2.plot(keys, prices, color='b') # de ander een lijngrafiek

# weer t zelfde als de voorheen
ax1.set_xlabel('Product Name') # gezamelijke as label
ax1.set_ylabel('Quantity Ordered', color='g') # label van as y voor de een,
ax2.set_ylabel('Price ($)', color='b') # label van as y voor de ander
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```

## 4 Welk product werd het meest verkocht? En waarom?

Betterijen. Waarschijnlijk omdat ze zo goedkoop zijn en vaak vervangen moeten worden/ veel in gebruik zijn.
